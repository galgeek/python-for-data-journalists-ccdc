Create a subset by filtering contribs using the *complete* committee name of the committee with by far the largest number of contributions.

```python
pocontribs = contribs[contribs['committee_name'] == 'CALIFORNIA CORRECTIONAL PEACE OFFICERS ASSOCIATION TRUTH IN AMERICAN GOVERNMENT FUND; NO ON PROPOSITION 62, YES ON PROPOSITION 66 (NON-PROFIT 501 (C) 5)']
```

```python
pocontribs.info()
```

There were *many* contributions to this committee of $287.08.

```python
pocontribs.describe()
```

```python
pocontribs['amount'].min()
```

```python
pocontribs[pocontribs['amount'] == pocontribs['amount'].min()]
```

Create another subset, filtering with other comittee data.

```python
prop = props[props.prop_name == 'PROPOSITION 058 - SB 1174 (CHAPTER 753, STATUTES OF 2014), LARA. ENGLISH LANGUAGE EDUCATION']
```

```python
prop.info()
```

```python
propcontribs = pd.merge(prop, contribs, on='calaccess_committee_id')
```

```python
propcontribs.info()
```

```python
propcontribs.amount.sum()
```

```python
propsupport = propcontribs[propcontribs.committee_position == 'SUPPORT']
```

```python
propoppose = propcontribs[propcontribs.committee_position == 'OPPOSE']
```

```python
propsupport.info()
```

```python
propoppose.info()
```

```python
propsupport.amount.sum()
```

```python
propoppose.amount.sum()
```

```python
round(propsupport.amount.sum() / propcontribs.amount.sum(), 2)
```

```python
from decimal import Decimal
```

```python
round(Decimal(propsupport.amount.sum() / propcontribs.amount.sum()), 2)
```

```python
propsupport.sort_values('amount', ascending=False).head()
```

```python
propoppose.sort_values('amount', ascending=False).head()
```

```python
propcontribs.groupby('committee_name_x').amount.sum().reset_index().sort_values('amount', ascending=False)
```

```python
propcontribs.groupby(['contributor_firstname','contributor_lastname','contributor_zip','committee_position']).amount.sum().reset_index().sort_values('amount', ascending=False)
```
